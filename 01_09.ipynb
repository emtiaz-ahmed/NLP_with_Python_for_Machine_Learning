## NLP Basics: Implementing a pipeline to clean text

#### Pre-processing text data.
1. Remove punctuation
2. Tokenization
3. Remove stopwords
4. Lemmatize/Stem

```python
import pandas as pd
## default 50, changed to 100
pd.set_option('display.max_colwidth', 100)
data = pd.read_csv("SMSSpamCollection.tsv", sep="\t", header=None)
data.columns = ['label', 'body_text']
data.head()
```

#### Remove Punctuation



```python
import string
## list of punctuations
string.punctuation
```

```python
## function for removing punctuations
def remove_punc(text):
    ## based on space it joins characters
    text_nopunc = "".join([char for char in text if char not in string.punctuation])
    return text_nopunc

data['body_text_clean'] = data['body_text'].apply(lambda x: remove_punc(x))

data.head()

```

#### Tokenization

```python
import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

## lower() converts texts into lower Case texts 
data['body_text_tokenize'] = data['body_text_clean'].apply(lambda x: tokenize(x.lower()))
data.head()
```

#### Remove stopwords

```python
import nltk

stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]
    return text

data['body_text_nostop'] = data['body_text_tokenize'].apply(lambda x: remove_stopwords(x)) 

data.head()
```

#### Stem text

#### -- it indenfies similar words. and set common word

```python

ps = nltk.PorterStemmer()

## list of PorterStemmer. and we will use the "Stem"
dir(ps)
```

```python
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

data['body_text_stemmed'] = data['body_text_nostop'].apply(lambda x: stemming(x))

data.head()
```

#### Lemmatizing
#### --- grouping together words based on word's lemma
#### --- better than Stemming . But takes more computation time.

#### Steps of WordNet lemmatizer

```python
wn = nltk.WordNetLemmatizer()
dir(wn)
```

```python
### commonly used "lemmatize". we will use it.

print(ps.stem('meanness'))
print(ps.stem('meaning'))
print(wn.lemmatize('meanness'))
print(wn.lemmatize('meaning'))
```

```python
## lemmatize text

def lemmatizing(tokenized_text):
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text

data['body_text_lemmatized'] = data['body_text_nostop'].apply(lambda x: lemmatizing(x))

data.head()
```
